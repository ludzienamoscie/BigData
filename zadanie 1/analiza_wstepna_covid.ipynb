```python
# 3.7.3 Dodatkowe dane: Liczba pastw, w ktrych liczba zmarych kobiet jest mniejsza od liczby zmarych mczyzn
q = """
with tab as (
    select country_name, sum(new_deceased_male), sum(new_deceased_female)
    from bigquery-public-data.covid19_open_data.covid19_open_data
    where aggregation_level = 0
    group by country_name
    having sum(new_deceased_male) > sum(new_deceased_female)
) select count(*) `no_countries` from tab
"""
query_job = client.query(q)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 3.7.4 Dodatkowe dane: Liczba pastw, w ktrych liczba zmarych mczyzn jest mniejsza od liczby zmarych kobiet
q = """
with tab as (
    select country_name, sum(new_deceased_male), sum(new_deceased_female)
    from bigquery-public-data.covid19_open_data.covid19_open_data
    where aggregation_level = 0
    group by country_name
    having sum(new_deceased_male) < sum(new_deceased_female)
) select count(*) `no_countries` from tab
"""
query_job = client.query(q)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 4.1 Chcemy posiada podstawowe dane o wszystkich krajach wiata, ktre bd zrozumiae dla czowieka oraz uniwersalne i potencjalnie przyszociowe do dalszego przetwarzania.
q = ("SELECT distinct country_name, population, area_sq_km, population_density FROM `bigquery-public-data.covid19_open_data.covid19_open_data` WHERE country_name IS NOT NULL AND aggregation_level = 0")
query_job = client.query(q)
query_result = query_job.result()

new_csv = query_result.to_dataframe().dropna().drop_duplicates()
new_csv.to_csv('zad4.1.csv', index=False)

new_csv
```

```python
# 4.2 Chcemy wygenerowa statystyki dotyczce zachorowa na COVID-19 na caym wiecie.
q = ("SELECT date, SUM(new_confirmed) AS confirmed_cases, SUM(new_deceased) AS deaths, SUM(new_recovered) AS recovered, SUM(new_confirmed) - SUM(new_deceased) - SUM(new_recovered) AS active_cases FROM `bigquery-public-data.covid19_open_data.covid19_open_data` WHERE aggregation_level = 0 GROUP BY date ORDER BY date")
query_job = client.query(q)
query_result = query_job.result()
new_csv = query_result.to_dataframe().dropna().drop_duplicates()
new_csv.to_csv('zad4.2.csv', index=False)

new_csv
```

```python
# 4.3 Chcemy pozna efekty COVID-19 poprzez uwypuklenie problemu miertelnoci ludzi spowodowanej wirusem.
q = ("SELECT date, SUM(new_deceased) AS total_deaths, SUM(new_confirmed) AS total_cases, (SUM(new_deceased)/SUM(new_confirmed))*100 AS case_fatality_rate FROM `bigquery-public-data.covid19_open_data.covid19_open_data` WHERE aggregation_level = 0 AND new_deceased IS NOT NULL AND new_confirmed IS NOT NULL AND date < '2022-09-15' GROUP BY date ORDER BY date")
query_job = client.query(q)
query_result = query_job.result()
new_csv = query_result.to_dataframe().dropna().drop_duplicates()
new_csv.to_csv('zad4.3.csv', index=False)

new_csv
```
